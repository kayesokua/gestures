```python
df.loc[:, 'torso_x'] = (df['left_shoulder_x'] + df['right_shoulder_x'] + df['left_hip_x'] + df['right_hip_x']) / 4
df.loc[:, 'torso_y'] = (df['left_shoulder_y'] + df['right_shoulder_y'] + df['left_hip_y'] + df['right_hip_y']) / 4
df.loc[:, 'torso_z'] = (df['left_shoulder_z'] + df['right_shoulder_z'] + df['left_hip_z'] + df['right_hip_z']) / 4
```

## Joint Trajectories

Visualizing the trajectories of each joint over time can provide valuable insights into the movement patterns and styles of a dance. By plotting these trajectories using 2D or 3D scatter or line plots, you can identify unique movements and gestures that are specific to certain dance styles. Additionally, heatmaps can be used to visualize the frequency of movements across different body parts, helping to identify patterns and trends.

When analyzing joint trajectories, it's important to consider the movement in all three dimensions.
1. X-Joint Trajectories display the horizontal movement, providing insights into how far the dancer travels.
2. Y-Joint Trajectories represent vertical movements, including standing, suspended, and floorwork.
3. Z-Joint Trajectories represent the distance between the subject and the camera, with negative values indicating stepping backward and positive values indicating forward movements.

```python
# Get joint columns
joints_x = ["nose_x","torso_x","left_shoulder_x","right_shoulder_x","left_wrist_x","right_wrist_x","left_hip_x","right_hip_x"]
joints_y = ["nose_y","torso_y","left_shoulder_y","right_shoulder_y","left_wrist_y","right_wrist_y","left_hip_y","right_hip_y"]
joints_z = ["nose_z","torso_z","left_shoulder_z","right_shoulder_z","left_wrist_z","right_wrist_z","left_hip_z","right_hip_z"]

colors = ['slateblue', 'indigo', 'blue', 'orange', 'purple']
line_styles = ['-', '--', '-.', ':', '--']

fig, ax = plt.subplots(figsize=(24, 6))
for i, (joint_x, joint_y, joint_z) in enumerate(zip(joints_x, joints_y, joints_z)):
    ax.plot(df["frame"], df[joint_x], color=colors[i % len(colors)], linestyle=line_styles[i % len(line_styles)], label=joint_x[:-2])
ax.set_xticks(df["frame"][::48]) # This represents every 2s
ax.set_xlim(0, len(df))
ax.set_xlabel('Time')
ax.set_ylabel('X Coordinate')
ax.set_title('X-Joint Trajectories')
ax.grid(True)
ax.legend()
plt.show()

fig, ax = plt.subplots(figsize=(24, 6))
for i, (joint_x, joint_y, joint_z) in enumerate(zip(joints_x, joints_y, joints_z)):
    ax.plot(df["frame"], df[joint_y], color=colors[i % len(colors)], linestyle=line_styles[i % len(line_styles)], label=joint_x[:-2])
ax.set_xticks(df["frame"][::48]) # This represents every 2s
ax.set_xlim(0, len(df))
ax.set_xlabel('Time')
ax.set_ylabel('Y Coordinate')
ax.set_title('Y-Joint Trajectories')
ax.grid(True)
plt.show()

fig, ax = plt.subplots(figsize=(24, 6))
for i, (joint_x, joint_y, joint_z) in enumerate(zip(joints_x, joints_y, joints_z)):
    ax.plot(df["frame"], df[joint_z], color=colors[i % len(colors)], linestyle=line_styles[i % len(line_styles)], label=joint_x[:-2])
ax.set_xticks(df["frame"][::48]) # This represents every 2s
ax.set_xlim(0, len(df))
ax.set_xlabel('Time')
ax.set_ylabel('Z-Joint Coordinate')
ax.set_title('Joint Trajectories')
ax.grid(True)
plt.show()
```

## Motion Energy Images

A motion energy pyramid provides a hierarchical representation of the motion energy in a video, capturing both spatial and temporal features at different scales, which can be used to analyze and compare motion patterns within or across videos.

* In the energy map generated by PyMoten, the color intensity represents the magnitude of motion energy at each location: yellow indicates high motion energy (more movement), and blue indicates low motion energy (less movement); the x-axis corresponds to time, while the y-axis represents the spatial or frequency domain depending on the implementation.
* High energy motion generally indicates more rapid or larger movements in the video. It's an aggregated measure of the speed and amplitude of the movements.
* Overall, this MotionEnergyPyramid instance uses a total of 2530 filters (3 temporal frequency bands * 5 spatial frequency bands * 8 motion directions * 21 spatial locations) to analyze the motion in the video across different spatial and temporal scales and directions.

https://link.springer.com/article/10.3758/s13428-021-01612-7
https://gallantlab.org/pymoten

```python
nimages = 200
stimulus_fps = 24
video_file = 'data/interim/frames/contemporary_dance_solo_20230219225812_10_annotated.mp4'
luminance_images = moten.io.video2luminance(video_file, nimages=nimages)
nimages, vdim, hdim = luminance_images.shape
pyramid = moten.pyramids.MotionEnergyPyramid(stimulus_vhsize=(vdim, hdim),
                                             stimulus_fps=stimulus_fps)
print(pyramid)

features = pyramid.project_stimulus(luminance_images)
print(features.shape)
fig, ax = plt.subplots(figsize=(24, 12))
ax.matshow(features, aspect='auto')
ax.set_xticks(df["frame"][::48]) # This represents every 2s
ax.set_xlabel('Time')
plt.show()
```

Each of the plot_spatial_gabors() functions you used visualizes spatial components of the Gabor filters with the specified indices. In the context of contemporary dance, these visualizations can provide insights into the spatial properties of motion patterns captured by the filters. Here's a breakdown of what each function reveals:

* plot_spatial_gabors(pyramid, [0, 23, 88, 577]): Visualizes filters that capture motion patterns with different spatial scales, locations, and directions. This can help you understand how the dancer's movements vary across the performance space and at various levels of detail.

* plot_spatial_gabors(pyramid, [46, 52, 58, 64]): Focuses on filters with slightly varying motion directions. This can help you identify if the dancer is favoring certain directions or if they are incorporating diverse movement directions into their performance.

* plot_spatial_gabors(pyramid, [88, 157, 202, 247]): Showcases filters capturing a range of motion characteristics in the contemporary dance, such as different spatial and temporal frequencies. This can help you analyze the overall complexity and variety of the dancer's movements.

Regarding the position and size of the filters:

* Right-most/left-most: Filters positioned on the right or left edges of the visualization indicate that the dancer's movements have a strong horizontal component, capturing motion patterns towards the right or left of the performance space.
* Top-most: Filters positioned at the top of the visualization indicate that the dancer's movements have a strong vertical component, capturing motion patterns towards the top of the performance space.
* Small filters: Smaller filters in the visualization represent higher spatial frequencies, capturing finer details of the dancer's movements, such as small and rapid gestures.


```python
def plot_spatial_gabors(pyramid, indices):
    vdim = pyramid.definition['vdim']

    size = 0.6
    fig, axs = plt.subplots(
        2, len(indices), figsize=((len(indices) * 4.25 + 0.25) * size,
                                  7.5 * size), squeeze=False,
        sharex=True, sharey=True)

    axs.T[0][0].set_ylabel('spatial Gabor (cos)')
    axs.T[0][1].set_ylabel('spatial Gabor (sin)')

    for ii, axs_ in zip(indices, axs.T):

        spatial_sin, spatial_cos = pyramid.get_filter_spatial_quadrature(ii)
        axs_[0].imshow(spatial_cos, aspect='equal')
        axs_[1].imshow(spatial_sin, aspect='equal')
        axs_[0].set_title('filter %d' % ii)

        gabor = pyramid.filters[ii]
        for ax in axs_:
            ax.scatter(gabor['centerh'] * vdim, gabor['centerv'] * vdim,
                       color='k', marker='+')
            ax.set_xticks([])
            ax.set_yticks([])

    plt.tight_layout()

# Different spatial frequencies
plot_spatial_gabors(pyramid, [0, 23, 88, 577])

# Different locations
plot_spatial_gabors(pyramid, [46, 52, 58, 64])

# Different directions
plot_spatial_gabors(pyramid, [88, 157, 202, 247])
```

## Principal Component Analysis

Principal Component Analysis (PCA) Visualization: PCA can be used to reduce the dimensionality of the kinematic data and project it onto a lower-dimensional space that can be easily visualized. This can help to identify the most important features that distinguish different dance styles or gestures.

```python
from sklearn.preprocessing import StandardScaler
kinematic_data = df[['nose_x', 'nose_y', 'nose_z','torso_x','torso_y','torso_z'].values

scaler = StandardScaler()
kinematic_data_standardized = scaler.fit_transform(kinematic_data)
pca = PCA(n_components=2)  # Reduce the data to 2 dimensions
principal_components = pca.fit_transform(kinematic_data_standardized)
plt.scatter(principal_components[:, 0], principal_components[:, 1])
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA Visualization of Dance Data')
plt.show()
```
